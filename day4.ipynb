```python
import streamlit as st
import joblib

model = joblib.load("netflix_type_model.pkl")
le = joblib.load("rating_encoder.pkl")

st.title("Netflix Content Type Predictor")

movie_age = st.number_input("Movie age (years)", min_value=0, max_value=100)
rating = st.selectbox("Rating", le.classes_)

if st.button("Predict"):
    rating_encoded = le.transform([rating])
    result = model.predict([[movie_age, rating_encoded[0]]])
    st.success(f"This is likely a: {result[0]}")
```
